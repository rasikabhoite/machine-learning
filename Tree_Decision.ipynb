```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
#Spliting dataset into indepenedent and dependent data 
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
#Splitting Decision tree regression model
from sklearn.tree import DecisionTreeRegressor,plot_tree
reg=DecisionTreeRegressor(random_state=0)
reg.fit(x,y)
```

```python
# Predict for my input
myinput=pd.DataFrame(data=[[4]],columns=["Level"])
reg.predict(myinput)
```

```python
# Visualize Decision Tree
plt.figure(figsize=(50,50))   #set size of figure
plot_tree(reg,feature_names=["Level"],filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
## Predict for user input
level=float(input("Enter Level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
result=reg.predict(myinput)
print(f"Predicated result for level{level}:{round(result[0])}")
```
