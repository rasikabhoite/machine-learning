```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset
ds=pd.read_csv("Salary_Data.csv")
ds
```

```python
#Divide dataset as independent and dependent variable
x=ds[['YearsExperience']]
y=ds[['Salary']]
```

```python
#plot
plt.scatter(x,y)
plt.show
```

```python
y
```

```python
#Splitting dataset as Training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train
```

```python
# Introduce Model
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
reg.fit(x_train,y_train)
```

```python
#Predict
result=reg.predict([[8]])
result
```

```python
result=reg.predict(x_test)
result
```

```python
y_test
```

```python
# Vislusal Plot
plt.scatter(x,y)
result=reg.predict(x)
plt.plot(x,result,color='green')
plt.show()
```

```python
# For user Input
years=int(input("Enter years of exp:"))
myinput=pd.DataFrame(columns=['YearsExperience'],data=[[years]])
result=reg.predict(myinput)
print("Your Salary must be close to",round(result[0,0]))
```
