```python
#Handling missing values (using mean and scikit learn)

from sklearn.impute import SimpleImputer
si=SimpleImputer(missing_values=np.nan,strategy='mean')
ds[['Age','Salary']]=si.fit_transform(ds[['Age','Salary']])
```

```python
ds
```

```python
# Encoding Categorical column (oneHotEncoder)
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse_output=False)
encode_data=ohe.fit_transform(ds[['Country']])
encode_data
```

```python
#creating colums
column_names=ohe.get_feature_names_out(['Country'])
encoded_df=pd.DataFrame(encode_data,columns= column_names)
encoded_df
```

```python
#Joining original dataset and new dataframe

ds = pd.concat([ds,encoded_df],axis=1)
ds
```

```python
#Drop actual categorical country column

ds = ds.drop(columns =['Country'])
ds
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Purchased']=le.fit_transform(ds['Purchased'])
```

```python
ds
```

```python
#Feature scaling 
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
ds[['Age','Salary']]=sc.fit_transform
```

```python
ds
```

```python
#Divied dataset into  Independent and Dependent
x=ds[['Age','Salary','Country_France','Country_Germany','Country_Spain']]
y=ds[['Purchased']]
```

```python
#Splitting datset as Training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0,random_state=1)
```
