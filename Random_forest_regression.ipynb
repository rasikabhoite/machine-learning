```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#import dataset and reading dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
#Spliting dataset into indepenedent and dependent data 
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
# Fitting Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=10,random_state=0)
reg.fit(x,y)
```

```python
# Predict for myinput
myinput=pd.DataFrame(data=[[4]],columns=['Level'])
reg.predict(myinput)
```

```python
# Visalization of random forest regression tree
from sklearn.tree import plot_tree
plot_tree(reg.estimators_[0],filled=True)
plt.show()
```

```python
# Predict for user input
level=float(input("Enter Level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])

result=reg.predict(myinput)
print(f"Predicated result for level{level}:₹{result[0]}")
```
